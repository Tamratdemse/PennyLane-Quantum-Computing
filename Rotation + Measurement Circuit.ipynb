# Importing the necessary libraries 

```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt
```

# 1-qubit device

```python
dev = qml.device("default.qubit", wires=1)
```

# Define the QNode with a variable angle for RX

```python
@qml.qnode(dev)
def rotation_circuit(theta):
    qml.RX(theta, wires=0)
    return qml.expval(qml.PauliZ(0))
```

# Create a range of angles from 0 to 2π

```python
angles = np.linspace(0, 2 * np.pi, 100)
expectation_values = [rotation_circuit(theta) for theta in angles]
```

# Plot the expectation value vs. angle

```python
plt.plot(angles, expectation_values)
plt.xlabel("Rotation angle θ (radians)")
plt.ylabel("Expectation value ⟨Z⟩")
plt.title("Expectation value of Pauli Z vs RX rotation angle")
plt.grid(True)
plt.show()
```
